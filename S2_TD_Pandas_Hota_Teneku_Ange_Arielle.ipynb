```python

import pandas as pd
import numpy as np
data = {
    'produit': ['Riz (1kg)', 'Huile (1L)', 'Tomate (kg)', 'Poulet (kg)',
                'Plantain (kg)', 'Pain (baguette)', 'Sucre (1kg)'],
    'marche': ['Mokolo', 'Sandaga', 'Mokolo', 'Marché Central',
                'Sandaga', 'Marché Central', 'Mokolo'],
    'prix_fcfa': [500, 1200, 250, 2500, 300, 200, 600],
    'stock_kg': [500, 200, 150, 80, 350, None, 420],
    'bio': [False, False, True, False, True, False, None],
}
df = pd.DataFrame(data)


# forme du dataFrame
print(df)

# les dimession du data frame
print(f'ce dataFrame a au total (lignes et colonnes) :\n', df.shape)

# Affichez le résumé complet avec df.info(). Quelles colonnes ont des valeurs manquantes ?
print(df.info())

#  Affichez les statistiques descriptives. Quel est le prix moyen ? Le prix médian ?
print(df.describe())

# Combien de produits sont disponibles dans chaque marché ? (utiliser value_counts)
print(f'le nombre de produits disponibles sur chaque marché est : \n', df.value_counts('marche'))

# Sélectionnez uniquement les colonnes 'produit' et 'prix_fcfa' avec iloc (en utilisant les positions).
print(df.iloc[:, [0,2]])

# Affichez tous les produits dont le prix est supérieur à 500 FCFA
print(f'les produits dont le prix est supérieur à 500 fcfa sont : \n', df[df['prix_fcfa']>500])

# Affichez les produits bio disponibles au marché de Mokolo
print(f'les produit bio disponoble au marché mokolo sont : \n', df[(df['bio']) & (df['marche'] == 'Mokolo')])

# Utilisez query() pour trouver les produits avec un stock supérieur à 300 kg ET un prix inférieur à
# 600 FCFA.

print(f'le produit avec un stock supérieur à 300 kg et au prix inférieur à 600 fcfa sont : \n', df.query('stock_kg > 300 and prix_fcfa < 600'))
```

```python
import pandas as pd
import numpy as np

ventes_sales = pd.DataFrame({
    'date': ['2024-01-15', '2024-01-15', '2024-01-16', '2024-01-16',
             '2024-01-17', '2024-01-15', None, '2024-01-18'],
    'vendeur': ['Alice', 'BOB', 'alice', 'Bob', 'ALICE', 'Bob',
                'Charlie', 'Charlie'],
    'produit': ['Riz', 'Huile', 'Riz', 'Tomate', 'Riz', 'Huile',
                'Sucre', 'Sucre'],
    'quantite': [10, 5, 8, 20, -3, 5, 15, 1500],
    'prix_unit': [500, 1200, 500, 250, 500, 1200, None, 600],
    'total': [5000, 6000, 4000, 5000, -1500, 6000, None, 900000],
})

print(ventes_sales)

# valeurs manquantes
print("Valeurs manquantes par colonne :\n", ventes_sales.isnull().sum())

# doublons
print("Doublons exacts :", ventes_sales.duplicated().sum())

# standardisation noms vendeurs
ventes_sales['vendeur'] = ventes_sales['vendeur'].str.title()

print(ventes_sales['vendeur'])

# supprimer lignes avec date manquante
ventes_sales = ventes_sales.dropna(subset=['date'])

print("Après suppression des dates manquantes :\n", ventes_sales)

# supprimer doublons (CORRIGÉ)
ventes_sales = ventes_sales.drop_duplicates()

print("Après suppression des doublons :\n", ventes_sales)

# imputation médiane pour Sucre
median_sucre = ventes_sales.loc[
    ventes_sales['produit'] == 'Sucre', 'prix_unit'
].median()

ventes_sales.loc[
    (ventes_sales['produit'] == 'Sucre') & (ventes_sales['prix_unit'].isna()),
    'prix_unit'
] = median_sucre

print("Après imputation médiane :\n", ventes_sales)
```
